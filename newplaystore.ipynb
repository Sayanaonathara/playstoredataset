1. Which app category has the highest average rating ?
top 10 categories by average rating.

```python
top_categories=avg_rating.head(10)
plt.figure(figsize=(8,8))
plt.pie(top_categories,
        labels=top_categories.index,
        autopct="%1.1f%%",
        startangle=140)
plt.title("Top 10 App Categories By Average Rating")
plt.show()
```

3. Are paid apps rated higher than free apps?

```python
avg_rating_by_type=df.groupby("Type")["Rating"].mean()
avg_rating_by_type.plot(kind="bar")
plt.title("Average Rating:Free vs Paid Apps")
plt.xlabel("App Type")
plt.ylabel("Average Rating")
plt.show()
```

Paid apps have a slightly higher average rating than free apps.

4. Do free apps get more installs than paid apps?

```python
plt.figure()
sns.barplot(x="Type",y="Installs",data=df)
```

yes,free apps get significantly more installs than paid apps.

5. Is there a relationship between the number of reviews and app rating? Do apps with more reviews tends to have better rating?

```python
import matplotlib.pyplot as plt
plt.scatter(df["Reviews"],df["Rating"])
plt.xlabel("Number of Reviews")
plt.ylabel("Rating")
plt.title("Relationship between Reviews and Rating")
plt.show()
```

There is no strong relation between reviews and rating. 
Apps with more reviews does not necessarily have better ratings.

6. Do more installs lead to more reviews?

```python
plt.scatter(df["Reviews"],df["Installs"])
plt.xlabel("Number of Reviews")
plt.ylabel("Installs")
plt.title("Relationship between Reviews and Installs")
plt.show()
```

Yes, more installs generally lead to more reviews,there is a positive relationship between installs and reviews.
Insights: 
Apps with higher installs tend to have more reviews.
Apps with low istalls usually have very few reviews.
Some apps have high installs but comparatively fewer reviews, meaning not all users leave reviews.
